```python
video_duration = 55
seconds_list = list(range(video_duration + 1))
seconds_list
empty_series = pd.Series(index=range(video_duration + 1), dtype=str)
pd_labels = pd.DataFrame(empty_series, columns=['label'])
pd_labels.index.name = 'seconds'
# pd_labels.iloc[0:2]='Hello'
# pd_labels
```

```python
# https://www.geeksforgeeks.org/read-json-file-using-python/
import json


# Open and read the JSON file
with open('mock.json', 'r') as file:
    data = json.load(file)

print(data)

# Print the data
for rangeLabel in data:
    label = rangeLabel['label']
    ranges_clip = rangeLabel['rangesClip']
    for range_clip in ranges_clip:
        start_range = range_clip['rStart']/100*video_duration
        end_range = range_clip['rEnd']/100*video_duration

        start_range = int(start_range)
        end_range = int(end_range)
        pd_labels.iloc[start_range:end_range+1] = label

pd_labels
```
